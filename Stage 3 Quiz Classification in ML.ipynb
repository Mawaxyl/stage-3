```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
data =pd.read_csv(r'C:\Users\AYA WAHEED\Downloads\Data_for_UCI_named.csv') 
data
```

```python
data.isna().sum()
```

```python
data.drop('stab', axis=1, inplace=True)
data
```

```python
x = data.drop(columns='stabf')
x
```

```python
y = data['stabf']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_train_data = scaler.fit_transform(x_train)
normalised_test_data = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
```

```python
rfc_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
rfc_prediction =rfc_model.predict(x_test)
print(accuracy_score(rfc_prediction, y_test))
```
